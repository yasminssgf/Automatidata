### **Tugas 2c. Memahami data - Menyelidiki variabel**

Urutkan dan interpretasikan tabel data untuk dua variabel: `trip_distance` dan `total_amount`.

**Jawab tiga pertanyaan berikut:**

**Pertanyaan 1:** Urutkan variabel pertama Anda (`trip_distance`) dari nilai maksimum ke minimum, apakah nilainya tampak normal?

**Contoh jawaban:**

**Pertanyaan 1:** Nilai-nilai tersebut sesuai dengan temuan data kami sebelumnya, di mana kami memperhatikan bahwa perjalanan terpanjang sekitar 33 mil.

**Pertanyaan 2:** Ya, dua nilai pertama jauh lebih tinggi daripada yang lainnya.

**Pertanyaan 3:** Perjalanan termahal belum tentu yang terpanjang.

```python
# ==> EXEMPLAR CODE and OUTPUT

# Sort the data by trip distance from maximum to minimum value

df_sort = df.sort_values(by=['trip_distance'],ascending=False)
df_sort.head(10)
```

```python
# ==> EXEMPLAR CODE and OUTPUT

# Sort the data by total amount and print the top 20 values
total_amount_sorted = df.sort_values(
    ['total_amount'], ascending=False)['total_amount']
total_amount_sorted.head(20)
```

```python
# ==> EXEMPLAR CODE and OUTPUT

# Sort the data by total amount and print the bottom 20 values
total_amount_sorted.tail(20)
```

```python
# ==> EXEMPLAR CODE and OUTPUT

# Berapa banyak dari setiap jenis pembayaran yang terwakili dalam data?
df['payment_type'].value_counts()
```

```python
# ==> CONTOH KODE DAN HASIL

# Berapakah rata-rata tip untuk perjalanan yang dibayar dengan kartu kredit?
avg_cc_tip = df[df['payment_type']==1]['tip_amount'].mean()
print('Avg. cc tip:', avg_cc_tip)

# Berapakah rata-rata tip untuk perjalanan yang dibayar tunai?
avg_cash_tip = df[df['payment_type']==2]['tip_amount'].mean()
print('Avg. cash tip:', avg_cash_tip)
```

```python
# ==> EXEMPLAR CODE and OUTPUT

# Berapa kali setiap ID vendor muncul dalam data?
df['VendorID'].value_counts()
```

```python
# ==> EXEMPLAR CODE and OUTPUT

# Berapakah rata-rata total pendapatan untuk setiap vendor?
df.groupby(['VendorID']).mean(numeric_only=True)[['total_amount']]
```

```python
# ==> EXEMPLAR CODE and OUTPUT

# Saring data hanya untuk pembayaran kartu kredit.
credit_card = df[df['payment_type']==1]

# Saring data khusus kartu kredit hanya untuk jumlah penumpang.
credit_card['passenger_count'].value_counts()
```

```python
# ==> EXEMPLAR CODE and OUTPUT

# Hitung jumlah tip rata-rata untuk setiap jumlah penumpang (hanya pembayaran kartu kredit).
credit_card.groupby(['passenger_count']).mean(numeric_only=True)[['tip_amount']]
```
