# **Lakukan A/B Testing**

**Pertanyaan penelitian untuk proyek data ini:** “Apakah ada hubungan antara jumlah total ongkos dan jenis pembayaran?”

### Imports dan data loading

Impor paket dan pustaka yang dibutuhkan untuk menghitung statistik deskriptif dan melakukan uji hipotesis.

```python
import pandas as pd
from scipy import stats
```

```python
taxi_data = pd.read_csv("2017_Yellow_Taxi_Trip_Data.csv", index_col = 0)
```

Secara umum, statistik deskriptif berguna karena memungkinkan untuk dengan cepat mengeksplorasi dan memahami sejumlah besar data. Dalam hal ini, menghitung statistik deskriptif membantu dengan cepat membandingkan rata-rata total biaya perjalanan di antara berbagai jenis pembayaran.

### Eksplorasi Data

Menggunakan statistik deskriptif untuk melakukan Analisis Data Eksplorasi (EDA).

**Catatan:** Dalam dataset, `payment_type` dikodekan dalam bilangan bulat:
* 1: Kartu kredit
* 2: Tunai
* 3: Tidak ada biaya
* 4: Sengketa
* 5: Tidak diketahui

```python
# Kode statistik deskriptif untuk EDA
taxi_data.describe(include='all')
```

Proyek ini tertarik pada hubungan antara jenis pembayaran dan jumlah ongkos yang dibayarkan pelanggan. Salah satu pendekatannya adalah dengan melihat jumlah ongkos rata-rata untuk setiap jenis pembayaran.

```python
taxi_data.groupby('payment_type')['fare_amount'].mean()
```

Berdasarkan rata-rata yang ditampilkan, tampaknya pelanggan yang membayar dengan kartu kredit cenderung membayar tarif yang lebih besar daripada pelanggan yang membayar tunai. Namun, perbedaan ini mungkin muncul dari pengambilan sampel acak, bukan karena perbedaan tarif yang sebenarnya. Untuk menilai apakah perbedaan tersebut signifikan secara statistik, perlu dilakukan uji hipotesis.

### Pengujian Hipotesis

**Hipotesis nol**: Tidak ada perbedaan tarif rata-rata antara pelanggan yang menggunakan kartu kredit dan pelanggan yang menggunakan uang tunai.

**Hipotesis alternatif**: Ada perbedaan tarif rata-rata antara pelanggan yang menggunakan kartu kredit dan pelanggan yang menggunakan uang tunai.

Tujuan pada langkah ini adalah untuk melakukan uji t dua sampel. Langkah-langkah untuk melakukan uji hipotesis:

1. Nyatakan hipotesis nol dan hipotesis alternatif
2. Pilih tingkat signifikansi
3. Temukan p-value
4. Tolak atau gagal menolak hipotesis nol

$H_0$: Tidak ada perbedaan rata-rata biaya perjalanan antara pelanggan yang menggunakan kartu kredit dan pelanggan yang menggunakan uang tunai.

$H_A$: Ada perbedaan rata-rata biaya perjalanan antara pelanggan yang menggunakan kartu kredit dan pelanggan yang menggunakan uang tunai.

Memilih 5% sebagai tingkat signifikansi dan melanjutkan dengan uji t dua sampel.

```python
#Hypothesis test, A/B test
#significance level

credit_card = taxi_data[taxi_data['payment_type'] == 1]['fare_amount']
cash = taxi_data[taxi_data['payment_type'] == 2]['fare_amount']
stats.ttest_ind(a=credit_card, b=cash, equal_var=False)
```

Karena p-value secara signifikan lebih kecil daripada tingkat signifikansi 5%, maka menolak hipotesis nol.

Disimpulkan bahwa terdapat perbedaan yang signifikan secara statistik dalam jumlah tarif rata-rata antara pelanggan yang menggunakan kartu kredit dan pelanggan yang menggunakan uang tunai.

**Kesimpulan**
1. Wawasan bisnis utama yang dapat ditarik dari pengujian hipotesis adalah bahwa mendorong pelanggan untuk membayar dengan kartu kredit dapat menghasilkan lebih banyak pendapatan bagi pengemudi taksi.

2. Pertimbangan proyek pengujian A/B testing ini kemungkinan tidak realistis dan proyek ini membutuhkan asumsi bahwa penumpang dipaksa untuk membayar dengan satu cara atau cara lain, dan bahwa setelah diberitahu tentang persyaratan ini, mereka selalu mematuhinya. Data tidak dikumpulkan dengan cara ini, sehingga asumsi harus dibuat untuk mengelompokkan entri data secara acak untuk melakukan A/B testing. Kumpulan data ini tidak memperhitungkan penjelasan lain yang mungkin. Misalnya, penumpang mungkin tidak membawa banyak uang tunai, sehingga lebih mudah untuk membayar perjalanan yang lebih panjang/jauh dengan kartu kredit.
